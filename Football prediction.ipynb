```python
#difference calculation to compare the performance
df['GD'] = df['G']-df['GA']
df['xGD'] = df['xG']-df['xGA']
df['NPxGD'] = df['NPxG']-df['NPxGA']
#outperform vs underform
df['COMPARE_GOAL'] = df['GD']-df['xGD']

df = df.sort_values(by='COMPARE_GOAL', ascending = False)
df 
```

```python
import matplotlib.pyplot as plt
df['COMPARE_GOAL'].plot(kind = 'barh', y =df['Team'])
plt.show()
```

```python
print(list(df.columns))
```

```python
cols = ['G', 'GA', 'xG', 'xGA', 'NPxG', 'NPxGA', 'GD', 'xGD', 'NPxGD']

for i in cols:
    df[i+'_pg']= (df[i])/(df['Games'])
df = df.sort_values(by='NPxGD_pg',ascending=False)
df
```

```python
average_goals = np.mean(df['xG_pg'])
average_goals
```

```python
#Match Fixture
fixture = pd.read_csv('/Users/devanshchawla/Downloads/python-for-fantasy-football-master/2 - Matchup Adjustment/epl_fixtures.csv')
fixture
```

```python
#Merging
df = df[['Team','xG_pg','xGA_pg']]
fixtures = pd.merge(fixture, df, left_on = 'Home_Team', right_on='Team')
fixtures = pd.merge(fixtures, df, left_on = 'Away_Team', right_on='Team')
fixtures = fixtures.drop(['Team_x','Team_y'],axis=1)

fixtures.columns = ['Home_Team','Away_Team','xG_pg_home','xGA_pg_home','xG_pg_away','xGA_pg_away'] 
fixtures
```

```python
#
fixtures_copy = fixtures.copy()
fixtures_copy['xG_adjusted_home_team'] = ((fixtures_copy['xG_pg_home'])*(fixtures_copy['xGA_pg_away']))/(average_goals)
fixtures_copy['xG_adjusted_away_team'] = ((fixtures_copy['xG_pg_away'])*(fixtures_copy['xGA_pg_home']))/(average_goals)
fixtures_copy       
```

```python
#ACCOUNT FOR HOME ADVANTAGE
fixtures_copy['xG_adjusted_home_team'] = fixtures_copy['xG_adjusted_home_team']*1.1
fixtures_copy['xG_adjusted_away_team'] = fixtures_copy['xG_adjusted_away_team']*0.87
fixtures_copy['home_result'] = fixtures_copy['xG_pg_home'] - fixtures_copy['xG_pg_away']
fixtures_copy['away_result'] = fixtures_copy['xGA_pg_home'] - fixtures_copy['xGA_pg_away']
fixtures_copy
```
